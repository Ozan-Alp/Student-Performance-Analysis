```python
pattern="(.*)(school)|(college)|(associate)|(r's degree).*"
pattern2="(.*)(school|college|associate|r's degree).*"
levels={"school":"Low","college":"Mid","associate":"Mid","r's degree":"High"}
#levels=["Low","Mid","Mid","High"]
def manipulate(x):
    result=re.match(pattern2,x).group(2)
    return levels[result]
    
```

```python
data["parental level of education"]=data["parental level of education"].apply(manipulate)
```

# H5- All scores are correlated. Ex: If one student is good at math, s/he is also good at reading and writing

```python
data["math_rank"]=data["math score"]
data["reading_rank"]=data["reading score"]
data["writing_rank"]=data["writing score"]
ranks=data[["math_rank","reading_rank","writing_rank"]]
ranks.corr()
#ranks.head(50)
#POSITIVELY CORRELATED, students all scores are propo
```

```python
data["math_rank"]=(data["math score"]-data["math score"].mean())/data["math score"].mean()
data["reading_rank"]=(data["reading score"]-data["reading score"].mean())/data["reading score"].mean()
data["writing_rank"]=(data["writing score"]-data["writing score"].mean())/data["writing score"].mean()
ranks=data[["math_rank","reading_rank","writing_rank"]]
ranks.corr()
#ranks.head(50)
```
